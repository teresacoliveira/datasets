```r
########################################################

library(ggplot2)
library(tidyverse)

df <- read.csv('Biolog Demo Data - PivotTable1.csv')

ggplot(data = df) +
  aes(x=Cell.line) +
  aes(y=Pyruvic.Acid....L.Malic.Acid.100.uM.) +
  geom_jitter(shape = 20, ) +
  aes(color=Cell.line) +
  geom_boxplot(fill=NA) +
  theme_light()

########################################################
```

#B- Demonstration without file uploading

Load file from an online location (e.g. github)

```r
library(tidyverse)

# 1. Define the URL where the CSV file can be found
url <- 'https://raw.githubusercontent.com/teresacoliveira/datasets/refs/heads/main/Biolog%20Demo%20Data%20-%20PivotTable1.csv'

# 2. Read the CSV
# Using stringsAsFactors = FALSE is a good habit for modern R
df <- read.csv(url, stringsAsFactors = FALSE)

# 3. View the first few rows to confirm you loaded the correct file
head(df)
```

Run the code produced in ggplotteR (https://huygens.science.uva.nl/ggPlotteR/)

```r
library(ggplot2)

ggplot(data = df) +
  aes(x=Cell.line) +
  aes(y=Pyruvic.Acid....L.Malic.Acid.100.uM.) +
  geom_jitter(shape = 20, ) +
  aes(color=Cell.line) +
  geom_boxplot(fill=NA) +
  theme_light()
```
